# ödev

    Kullanıcıdan girdi olarak alınan rakam notun harf karşılığını ekrana yazdıran bir program yazınız.

    90-100 arası A
    80-89 arası B
    70-79 arası C
    60-69 arası D
    50-59 arası E
    0-49 arası F

# Cevap

```python
harf_not = ""
ders_notu = input("Notu giriniz: ")
try:
    ders_notu2 = int(ders_notu)
    if ders_notu2 > 100 or ders_notu2 <0:
        harf_not="Lütfen 0-100 arası bir rakam giriniz."
    elif ders_notu2 <= 100:
        harf_not = "A"
    elif ders_notu2 < 90:
        harf_not = "B"
    elif ders_notu2 < 80:
        harf_not = "C"
    elif ders_notu2 < 70:
        harf_not = "D"
    elif ders_notu2 < 60:
        harf_not = "E"
    else: harf_not = "F"  
    
    print(harf_not)
except ValueError:
    print("Lütfen sadece rakam giriniz.")
```

# Kimlik İşleci

    Python’da her şeyin (ya da başka bir deyişle her nesnenin) bir kimlik numarası (identity) vardır. Kabaca söylemek gerekirse, bu kimlik numarası denen şey esasında o nesnenin bellekteki adresini gösterir.

    Peki bir nesnenin kimlik numarasına nasıl ulaşırız?

```python
istanbul = 34
id(istanbul)
```

```python
ankara = 34
id(ankara)
```

    Python iki adet 100 sayısı için bellekte iki farklı nesne oluşturmuyor. İlk kullanımda önbelleğine aldığı sayıyı, ikinci kez ihtiyaç olduğunda bellekten alıp kullanıyor. Bu tür bir önbellekleme mekanizmasının gerekçesi performansı artırmaktır.
