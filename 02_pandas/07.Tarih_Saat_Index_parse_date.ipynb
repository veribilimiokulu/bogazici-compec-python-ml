# Ayların İsimlerine Göre Toplam Satış

```python
# Ayların isimleri ile rakamları eşleştiren fonksiyon
def month_switch(i):
        switcher={
                
                1:'Ocak',
                2:'Şubat',
                3:'Mart',
                4:'Nisan',
                5:'Mayıs',
                6:'Haziran',
                7:'Temmuz',
                8:'Ağustos',
                9:'Eylül',
                10:'Ekim',
                11:'Kasım',
                12:'Aralık'
             }
        return switcher.get(i)
```

```python
# Test
month_switch(3)
```

```python
# Pandas dataframe month series içinde dolaşarak paterne uygun eşleştirme yapalım
months = []
for i in df.month:
    months.append(month_switch(i))
```

```python
# Oluşturduğumuz paterni taşıyan listeyi yeni bir sütun olarak dataframe e ekleyelim
df["month_names"] = months
```

```python

def is_date_between_them(start_dates, end_dates, check_date):
    # bir tarihin iki tarih arasında olup olmadığını kontrol eden fonksiyon
    # Parametrelerde start_dates ve end_dates aynı dataframe e ait aralarında tarih aralığı bulunan tarihler
    flag = False
    # tarih aralıklarının içinde dolaşalım ve belirtilen tarihin aralıkta olup olmadığını kontrol edelim
    for start_date, end_date in zip(start_dates, end_dates):
        # tarih aralıkta ise flag true olsun değilse false
        if( (check_date >= start_date) and (check_date <= end_date) ):
            flag = True
        else: flag = False
    return flag
```

```python
# df e bir nitelik ekleyeceğimiz için onun uzunluğundaki bir listeyi paterne uygun dolduralım
is_promotion = []
# date sütunu içinde dolaş
for check_date in df.date:
    # eğer bu satırdaki tarih promotion tablosundaki tarihler arasında ise listeye 1 ekle yoksa 0
    if(is_date_between_them(start_dates=df_promotion.start_date, end_dates=df_promotion.end_date,check_date=check_date)):
        is_promotion.append(1)
    else: is_promotion.append(0)
```
