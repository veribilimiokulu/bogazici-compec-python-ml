```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dataset = pd.read_csv("D:/OneDrive/egitim/verilen/bogazici-compec-python-ml/Datasets/PozisyonSeviyeMaas.csv")
```

# Veri Keşfi

```python
dataset.head()
```

```python
plt.figure(figsize=(9,6))
plt.scatter(dataset['Seviye'], dataset['Maas'], color='blue')
plt.title("Pozisyon maaş")
plt.xlabel("Pozisyon")
plt.ylabel("Maaş")
plt.show()
```

# Veri setini X ve y olarak ayırma (Dikey bölme)

```python
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
print(X.shape)
print(y.shape)
```

# Polinom Regresyon için Ön Hazırlık

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
```

```python
X_poly.shape
```

```python
X_poly
```

# Polinom Lineer Regresyon ile Modeli Eğitmek

```python
from sklearn.linear_model import LinearRegression
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

# Polinom Regresyon Modelin Grafiğini Çizmek

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Pozisyon&Maas Polinom Regresyon')
plt.xlabel('Pozisyon')
plt.ylabel('Maas')
plt.show()
```

# Tahmin Yapma

```python
# Tahmin edilecek veri hazırlığı 
deger = 6.7
tahmin_edilecek_pozisyon = poly_reg.fit_transform(np.array(deger).reshape(-1, 1))
```
